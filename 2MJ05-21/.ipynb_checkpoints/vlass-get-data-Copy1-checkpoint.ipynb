```python
file = open('interesting_sources_usno_2.txt','r')

#print(file.readlines(1))

coords= []
alldict = dict()

for line in file:
  
  #Let's split the line into an array called "fields" using the ";" as a separator:
    fields = line.split(" ")
  
  #and let's extract the data:
#   coords.extend('\"'+fields[0]+':'+fields[1]+':'+fields[2]+'\", \"'+fields[3]+':'+fields[4]+':'+fields[5]+'\"')
    RA = fields[0]+':'+fields[1]+':'+fields[2]
    Dec = fields[3]+':'+fields[4]+':'+fields[5]
    try:
        loc= coord.SkyCoord(RA, Dec, unit=(u.hour, u.deg))
        print(str(loc) + ' VLASS1.1')
    #    cod = coord.SkyCoord(coord, unit=(u.hour, u.deg))
        tilename1 = get_vlass_1.get_tilename(loc)
    except AssertionError:
        pass
    try:
        loc= coord.SkyCoord(RA, Dec, unit=(u.hour, u.deg))
        print(str(loc) + ' VLASS1.2')
    #    cod = coord.SkyCoord(coord, unit=(u.hour, u.deg))
        tilename2 = get_vlass_2.get_tilename(loc)
    except AssertionError:
        print('File at '+ loc + ' not found' )

    try:
        get_vlass_1.get_fits(loc)
        x, fitsname = get_vlass_1.get_fitsname(loc)
        alldict[tilename1] = {RA,Dec}
        with open('interestingsources_2.txt', 'a') as f:
            f.write(RA + ' '+ Dec + ' ' + fitsname +' '+'\n')
    except AssertionError as e:
        pass
    try:
        get_vlass_2.get_fits(loc)
        y, fitsname2 = get_vlass_2.get_fitsname(loc)
        alldict[tilename2] = {RA,Dec}
        with open('interestingsources_2.txt', 'a') as f1:
            f1.write(RA + ' '+ Dec + ' '+ fitsname2 +' '+'\n')
    except AssertionError as e:
        print('File not found: ' + RA + ' ' + Dec)
        with open('failed2findfits.txt', 'a') as fail:
            fail.write(RA + ' '+ Dec + ' '+ '\n')
        


```
